14. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в
діапазоні від -3 до 3 та знайдіть суму елементів на головній діагоналі.

```python
matrix = np.random.uniform(-3, 3, (3,3))
diag_sum = np.trace(matrix)

diag_sum
```

15. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в
діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить суму елементів
кожного стовпця матриці.

```python
matrix = np.random.uniform(-0.5, 0.5, (3,3))
col_sums = matrix.sum(axis=0).reshape(-1,1)

col_sums
```
